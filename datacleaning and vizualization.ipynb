# Data vizualisation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv("./Data sets/Universities data.csv")
```

```python
data.head()
```

```python
data.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
data
```

```python
sns.histplot(data["Fees in Rupees"], bins=30,color="slateblue")
plt.title("count on Fees prices")
plt.show()
```

## observation
- Most of the colleges collecting around 1 lakh fee.

```python
sns.boxplot(y=data["Fees in Rupees"],palette='Set2')
plt.title("Boxplot of College Fees")
plt.show()
```

## observations
- College Fee in range of 20 thousand to 15 lakh.
- Most of the colleges Fee in the of 50 thousand to 6.5 lakh.
- Average College Fee is around 2.5 lakh.

```python
sns.histplot(data["Rating for 5"],kde=True)
plt.title("Kde plot on Ratings")
plt.show()
```

## observation
- Most college ratings are between 3.5 and 4.5.
- Colleges with 4 rating are more in number.

```python
plt.figure(figsize=(10,7))
ax = sns.countplot(y="State", data=data,color='orange')
plt.title("Number of Colleges by State")
for container in ax.containers:
    ax.bar_label(container, padding=3)
plt.show()
```

## observations
- Maharashtra, Delhi, Tamilnadu and karnataka has morethan 200 colleges in each.
- Uttar Pradesh, West Bengal, Rajasthan, Telangana, Gujarat, Kerala, Madhya Pradesh, Punjab and Andhra Pradesh contains morethan 50 colleges in each.
- The remaining states contains lessthan 50 colleges each.

```python
data.columns
```

```python
data["Degree"].value_counts().plot.pie(autopct="%1.1f%%", explode=[0.4, 0, 0, 0.3, 0, 0, 0])
plt.title("Approval Types")
plt.show()
```

## observation
- **BE/B.Tech** degress are higher compare to other degress.
- **BCA** degress are lower compare to other degrees.

```python
sns.countplot(
    y="City",
    data=data,
    order=data["City"].value_counts().head(10).index,
    color='orange'
)

plt.title("Top 10 Cities with Most Colleges")
plt.xlabel("Number of Colleges")
plt.ylabel("City")
plt.show()
```

## observation
- **New Delhi city** has more number of colleges compare to other city colleges.

```python
sns.countplot(
    y="Course",
    data=data,
    order=data["Course"].value_counts().head(10).index,
    palette='Set2'
)

plt.title("Top 10 Courses")
plt.xlabel("Count")
plt.ylabel("Courses")
plt.show()
```

## observation
- **BCA** and **B.Tech computer Science And Engineering** are more compare to other courses.

# Bivariate

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x="Fees in Rupees",
    y="Rating for 5",
    hue="Fees in Rupees",
    data=data
)

plt.title("Fees vs College Rating")
plt.xlabel("Fees")
plt.ylabel("Rating")
plt.show()
```

## observation
- Most of the ratings between 3.3 and 4.8 rating and fees in the range of thousand to 14 lakh.
- College ratings are not affecting on college Fee.

```python
top5_states = data["State"].value_counts().head().index

sns.countplot(
    y="State",
    hue="Degree",
    data=data[data["State"].isin(top5_states)]
)

plt.title("Degree Distribution by Top 5 States")
plt.show()
```

## Observation
- Different degree's provided by top 5 states.

```python
plt.figure(figsize=(13,5))
sns.boxplot(x="State", y="Fees in Rupees", data=data,showfliers=False,palette='Set2')
plt.xticks(rotation=90)
plt.title("Fees Distribution by State")
plt.show()
```

```python
plt.figure(figsize=(13,5))
sns.violinplot(x="Degree", y="Fees in Rupees", data=data)
plt.xticks(rotation=90)
plt.title("Rating Distribution by Course")
plt.show()
```

## Observation
